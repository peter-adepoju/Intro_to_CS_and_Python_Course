## Section 2: Types

Every value in Python has a **type**. The type determines what operations
are valid. Use `type()` to inspect any value.

```python
# Integer type
type(42)
```

```python
# Float type
type(3.14)
```

```python
# Note: dividing two integers gives a FLOAT
type(10 / 2)
```

```python
# But floor-dividing gives an INT
type(10 // 2)
```

```python
# The floating-point precision issue — this is NOT a bug!
0.1 + 0.2
```

```python
# Type conversion functions
print(int(3.9))    # truncates toward zero
print(int(3.1))    # truncates — does NOT round
print(round(3.9))  # rounds to nearest integer
print(float(5))    # converts int to float
print(int("42"))   # converts string to int
```

## Section 3: Variables

A variable is a name that **refers to** a value. Use `=` to create or
update a binding. Remember: `=` is assignment, not equality!

```python
# Create a variable
pi = 355 / 113
print(pi)
```

```python
# Use variables in expressions
radius = 2.2
area = pi * (radius ** 2)
circumference = 2 * pi * radius

print("Area:", area)
print("Circumference:", circumference)
```

```python
# Reassigning a variable
x = 5
print("x is", x)
x = 10
print("x is now", x)
```

```python
# Updating based on current value
count = 0
print("count starts at:", count)
count = count + 1
print("count after +1:", count)
count += 1  # shorthand for count = count + 1
print("count after another +1:", count)
```

### Guided Exercise: The Broken Swap

A common beginner mistake is trying to swap two variables incorrectly.
Trace through the following code to understand WHY it's broken,
then fix it.

```python
# BROKEN swap — trace through this line by line
x = 1
y = 2
print(f"Before: x={x}, y={y}")

# BUG: this doesn't actually swap!
y = x      # y is now 1 (x's value), but the original y=2 is LOST!
x = y      # x is now 1 (y's value), but y is already 1...

print(f"After (broken): x={x}, y={y}")  # Both will be 1!
```

```python
# FIX 1: Use a temporary variable
x = 1
y = 2
print(f"Before: x={x}, y={y}")

temp = x   # save x's value
x = y      # x gets y's value
y = temp   # y gets x's original value

print(f"After (fixed): x={x}, y={y}")
```

```python
# FIX 2: Python's elegant tuple swap (preview — we'll study tuples in Week 5)
x = 1
y = 2
print(f"Before: x={x}, y={y}")

x, y = y, x   # Python evaluates the right side FIRST, then assigns

print(f"After (elegant): x={x}, y={y}")
```

## Section 4: A Real Program

Let's combine what we've learned to write a real program.
This one computes how long it would take light to travel from
Earth to various celestial bodies.

```python
# Speed of light in km/s
speed_of_light = 299792  # km per second

# Distances in km
distance_moon = 384400
distance_mars = 225000000     # average
distance_sun = 149600000

# Time = distance / speed (in seconds)
time_moon = distance_moon / speed_of_light
time_mars = distance_mars / speed_of_light
time_sun = distance_sun / speed_of_light

# Convert seconds to minutes
time_moon_min = time_moon / 60
time_mars_min = time_mars / 60
time_sun_min = time_sun / 60

print("Travel time for light:")
print(f"  To the Moon:  {time_moon:.2f} seconds ({time_moon_min:.2f} minutes)")
print(f"  To the Sun:   {time_sun:.2f} seconds ({time_sun_min:.2f} minutes)")
print(f"  To Mars:      {time_mars:.2f} seconds ({time_mars_min:.2f} minutes)")
```

## Day 1 Wrap-Up

**Key things to remember from today:**

1. `int`, `float`, `str`, `bool` are Python's basic types — use `type()` to check.
2. `/` always gives a float; `//` gives integer (floor) division.
3. `%` gives the remainder — essential for divisibility checks.
4. Variables are names that point to values. `=` creates/updates the binding.
5. Updating a variable: `x = x + 1` or `x += 1`.
6. `int()` truncates (doesn't round). `round()` rounds to nearest.
