## Section 1: Variables as Labels, Not Boxes

The mental model matters. A variable is a **name pointing to a value**,
not a box containing a value. This distinction becomes crucial in Week 5
when we study mutable data. For now, let's see the implication:

```python
# When you bind two names to the same value:
a = 5
b = a    # b points to the SAME value as a

print(f"a = {a}, b = {b}")

# Now reassign a:
a = 10   # a now points to 10; b still points to 5

print(f"After reassigning a:")
print(f"a = {a}, b = {b}")  # b is unchanged!
```

```python
# The right side is evaluated BEFORE the assignment happens.
# This is crucial for understanding self-update patterns:

x = 5
print(f"x starts at {x}")

x = x + 3   # Step 1: evaluate x + 3 = 5 + 3 = 8
             # Step 2: bind x to 8
print(f"After x = x + 3: x = {x}")

x = x * x   # Step 1: evaluate x * x = 8 * 8 = 64
             # Step 2: bind x to 64
print(f"After x = x * x: x = {x}")
```

---

## Section 2: Augmented Assignment Operators

Python provides shorthand for common update patterns.

```python
# All augmented assignment operators
x = 20
print(f"Starting value: {x}")

x += 5
print(f"After x += 5:   {x}")   # 25

x -= 3
print(f"After x -= 3:   {x}")   # 22

x *= 2
print(f"After x *= 2:   {x}")   # 44

x //= 5
print(f"After x //= 5:  {x}")   # 8

x **= 2
print(f"After x **= 2:  {x}")   # 64

x %= 10
print(f"After x %= 10:  {x}")   # 4
```

## Section 3: Variable Naming

Good names are not just aesthetic, they make code readable and debuggable.

```python
# POOR names: What does this program do?
a = 1200
b = 0.045
c = 12
n = 360
r = a * (b/c) * (1 + b/c)**n / ((1 + b/c)**n - 1)
print(r)
```

```python
# GOOD names: Same calculation — now you can read it!
loan_principal = 1200
annual_interest_rate = 0.045
payments_per_year = 12
total_payments = 360

monthly_rate = annual_interest_rate / payments_per_year
monthly_payment = (
    loan_principal * monthly_rate * (1 + monthly_rate)**total_payments
    / ((1 + monthly_rate)**total_payments - 1)
)

print(f"Monthly mortgage payment: ${monthly_payment:.2f}")
```

## Section 5: Expressions and Statements

An **expression** evaluates to a value. A **statement** is an instruction.
In Jupyter, the last expression in a cell is displayed automatically.

```python
# Expression — produces a value, Jupyter displays it
3 + 4
```

```python
# Statement — does something but produces nothing to display
x = 7   # no output — this is a statement
```

```python
# Both in one cell: only the last expression displays
a = 3
b = 4
a + b   # this IS displayed (last expression)
```

```python
# Use print() when you want to display multiple values:
a = 3
b = 4
print(a)
print(b)
print(a + b)
```

## Section 6: Guided Practice

**Problem:** A bank offers a simple interest savings account.
If you deposit `principal` dollars at an annual rate of `rate`,
after `years` years you have `principal * (1 + rate * years)` dollars.

Write code that:
1. Defines the principal, rate, and number of years as variables
2. Computes the final amount
3. Computes the interest earned (final - principal)
4. Prints a nicely formatted summary

Use: principal = 1000, rate = 0.05 (5%), years = 10

```python
# Guided: follow the steps above

# Step 1: Define variables
principal = 1000
rate = 0.05
years = 10

# Step 2: Compute final amount
final_amount = principal * (1 + rate * years)

# Step 3: Compute interest earned
interest_earned = final_amount - principal

# Step 4: Print summary
print("Simple Interest Calculation")
print("=" * 30)
print(f"Principal:       ${principal:,.2f}")
print(f"Annual rate:     {rate*100:.1f}%")
print(f"Years:           {years}")
print(f"Final amount:    ${final_amount:,.2f}")
print(f"Interest earned: ${interest_earned:,.2f}")
```
