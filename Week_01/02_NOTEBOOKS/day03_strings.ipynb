## Section 1: Creating Strings

```python
# Three ways to delimit strings
s1 = 'single quotes'
s2 = "double quotes"
s3 = """triple quotes
can span
multiple lines"""

print(s1)
print(s2)
print(s3)
```

```python
# Why both single and double quotes?
# Use whichever avoids needing a backslash:
print("It's a beautiful day")   # apostrophe inside double quotes - fine
print('She said "hello"')        # quotes inside single quotes - fine
print('It\'s also fine')         # backslash escape - works but uglier
```

```python
# The empty string
empty = ""
print(len(empty))     # 0
print(empty + "hi")   # hi
```

## Section 2: String Operations

```python
# Concatenation (+)
first = "Hello"
second = "World"
print(first + ", " + second + "!")
```

```python
# Repetition (*)
print("ha" * 3)
print("-" * 40)
print("* " * 5 + "*")
```

```python
# len() counts all characters including spaces
print(len("hello"))     # 5
print(len("hi there"))  # 8 (space counts!)
print(len(""))           # 0
```

```python
# Membership test with 'in'
s = "python programming"
print("python" in s)    # True
print("java" in s)      # False
print("P" in s)         # False — case sensitive!
```

## Section 3: Indexing

```python
# Positive indexing (0-based)
s = "Python"
#    P  y  t  h  o  n
#    0  1  2  3  4  5

print(s[0])   # P — first character
print(s[1])   # y
print(s[5])   # n — last character
```

```python
# Negative indexing (count from the end)
s = "Python"
#    P   y   t   h   o   n
#   -6  -5  -4  -3  -2  -1

print(s[-1])   # n — last character
print(s[-2])   # o
print(s[-6])   # P — same as s[0]
```

## Section 4: Slicing

Syntax: `s[start : stop : step]`
- **start**: first index (inclusive, default = 0)
- **stop**: end index (**exclusive**, default = len(s))
- **step**: stride (default = 1)

```python
s = "abcdefgh"

# Basic slices
print(s[0:3])    # abc  (indices 0, 1, 2)
print(s[3:6])    # def  (indices 3, 4, 5)
print(s[5:8])    # fgh  (indices 5, 6, 7)
print(s[0:1])    # a    (just one character)
```

```python
# Omitting start or stop uses defaults
s = "abcdefgh"
print(s[:3])     # abc   (start defaults to 0)
print(s[5:])     # fgh   (stop defaults to end)
print(s[:])      # abcdefgh (entire string)
print(s[:-1])    # abcdefg  (all but last)
```

```python
# Slicing with step
s = "abcdefgh"
print(s[::2])    # aceg  (every 2nd character)
print(s[1::2])   # bdfh  (every 2nd, starting from index 1)
print(s[::-1])   # hgfedcba (step -1 reverses the string!)
print(s[7:1:-1]) # hgfedc (backwards, from 7 to 2)
```

```python
# Slicing doesn't raise IndexError — it clips
s = "hello"
print(s[0:100])   # hello (clips to end)
print(s[10:20])   # ''    (empty, start is past end)
```

## Section 5: String Immutability

```python
# Strings cannot be modified in place:
s = "hello"
try:
    s[0] = 'H'   # This is not allowed!
except TypeError as e:
    print(f"TypeError: {e}")
```

```python
# To 'modify' a string, create a new one:
s = "hello"
s = "H" + s[1:]    # Create new string; rebind s to it
print(s)            # Hello
```

```python
# Practical: build a string gradually
result = ""
result = result + "Hello"
result += ", "
result += "World"
result += "!"
print(result)
```
