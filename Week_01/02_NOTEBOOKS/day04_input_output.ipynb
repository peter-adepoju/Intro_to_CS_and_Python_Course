## Section 1: The `print()` Function

```python
# Basic print
print("Hello, world!")
```

```python
# Multiple arguments — separated by space by default
name = "Alice"
age = 30
score = 95.5
print("Name:", name, "Age:", age, "Score:", score)
```

```python
# Changing the separator with sep=
print("a", "b", "c")             # a b c (default)
print("a", "b", "c", sep="")     # abc
print("a", "b", "c", sep="-")    # a-b-c
print("a", "b", "c", sep=", ")   # a, b, c
```

```python
# Changing the end character with end=
print("Hello", end=" ")
print("World", end="!\n")
print()  # empty print adds a blank line
```

```python
# Escape characters
print("Line 1\nLine 2\nLine 3")   # \n is newline
print("Name\tAge\tScore")          # \t is tab
print('She said \'hello\'')        # \' is literal single quote
print("Path: C:\\Users\\Alice")    # \\ is literal backslash
```

## Section 2: The `input()` Function

**Note:** In Jupyter, `input()` opens a text box. Type your response
and press Enter. For automated notebooks, input() can be tricky —
for standalone programs, use a `.py` file with `python myfile.py`.

```python
# input() always returns a STRING
# Try it — type your name when prompted
name = input("What is your name? ")
print(f"Hello, {name}! Great to meet you.")
```

```python
# CRITICAL: Converting input to numbers
# Try entering the number 5 when prompted
user_input = input("Enter a number: ")
print(f"Type of input: {type(user_input)}")   # str!

# Convert to work with numbers:
number = int(user_input)
print(f"Doubled: {number * 2}")
```

```python
# Compact: convert on the same line
age = int(input("How old are you? "))
birth_year = 2024 - age   # approximate
print(f"You were probably born around {birth_year}.")
```

## Section 3: f-Strings

```python
# Basic f-string
name = "Alice"
age = 30
print(f"My name is {name} and I am {age} years old.")
```

```python
# Expressions inside {}
x = 7
print(f"{x} squared is {x**2}")
print(f"{x} * {x} = {x*x}")
print(f"Type of x: {type(x)}")
```

```python
# Format specifiers: :.Nf for decimal places
pi = 3.14159265358979
print(f"Pi to 0 decimals: {pi:.0f}")
print(f"Pi to 2 decimals: {pi:.2f}")
print(f"Pi to 5 decimals: {pi:.5f}")
```

```python
# Format specifiers: , for thousands separator
population = 8045311447
gdp = 105435.7
print(f"World population: {population:,}")
print(f"World GDP (billions): ${gdp:,.1f}")
```

```python
# Format specifiers: % for percentage
correct = 17
total = 20
ratio = correct / total
print(f"Score: {correct}/{total} = {ratio:.1%}")
```

```python
# Format specifiers: alignment in columns
# This is very useful for tables
print(f"{'Name':<10} {'Score':>6} {'Grade':^6}")
print("-" * 25)
print(f"{'Alice':<10} {95:>6} {'A':^6}")
print(f"{'Bob':<10} {82:>6} {'B':^6}")
print(f"{'Catherine':<10} {78:>6} {'C+':^6}")
```
