## Section 1: Boolean Values and Comparisons

```python
# Booleans: only two values
print(True)
print(False)
print(type(True))    # bool
```

```python
# Comparison operators — all return bool
print(5 == 5)    # True
print(5 == 6)    # False
print(5 != 6)    # True
print(5 < 6)     # True
print(5 > 6)     # False
print(5 <= 5)    # True
print(5 >= 6)    # False
```

```python
# String comparisons — alphabetical order
print("apple" == "apple")    # True
print("apple" < "banana")    # True (a < b alphabetically)
print("Apple" == "apple")    # False! Case-sensitive
print("Z" < "a")             # True — uppercase letters come before lowercase in ASCII
```

```python
# You can store booleans in variables
is_raining = True
has_umbrella = False
print(f"Raining: {is_raining}")
print(f"Has umbrella: {has_umbrella}")

# Boolean from computation
x = 42
is_even = (x % 2 == 0)
is_positive = (x > 0)
print(f"{x} is even: {is_even}")
print(f"{x} is positive: {is_positive}")
```

## Section 2: Boolean Operators: `and`, `or`, `not`

```python
# 'and' — both must be True
print(True and True)     # True
print(True and False)    # False
print(False and True)    # False
print(False and False)   # False
```

```python
# 'or' — at least one must be True
print(True or True)      # True
print(True or False)     # True
print(False or True)     # True
print(False or False)    # False
```

```python
# 'not' — flips the boolean
print(not True)    # False
print(not False)   # True
print(not (5 > 3)) # False (because 5 > 3 is True)
```

```python
# Real-world example: checking a range
age = 25
is_working_age = age >= 18 and age <= 65
print(f"Age {age} is working age: {is_working_age}")

# Python also allows chained comparisons:
is_working_age = 18 <= age <= 65   # Pythonic!
print(f"Age {age} is working age (chain): {is_working_age}")
```

```python
# Predict: True or False?
x = 10
y = 3

# Try to predict before running
print(x > 5 and y < 10)       # Prediction: ___
print(x < 5 or y < 10)        # Prediction: ___
print(not (x == 10))           # Prediction: ___
print(x > 0 and y > 0 and x + y == 13)  # Prediction: ___
```

## Section 3: The `if` Statement

```python
# Basic if
temperature = 35

if temperature > 30:
    print("It's hot today!")
    print("Stay hydrated.")

print("This always prints.")  # Outside the if block
```

```python
# if-else: one or the other always runs
number = int(input("Enter a number: "))

if number % 2 == 0:
    print(f"{number} is even.")
else:
    print(f"{number} is odd.")
```

```python
# if-elif-else: for multiple exclusive conditions
score = int(input("Enter your score (0-100): "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
```

## Section 5: Common Mistakes in Branching

```python
# MISTAKE 1: Using if when you should use elif
score = 95

# WRONG — all if's are independent!
print("=== WRONG ===")
if score >= 90:
    print("A")    # Prints A
if score >= 80:   # Still checked! score=95 passes this too
    print("B")    # Also prints B!
if score >= 70:
    print("C")    # Also prints C!

print("\n=== CORRECT ===")
# CORRECT — use elif
if score >= 90:
    print("A")    # Prints A
elif score >= 80: # NOT checked because previous if was True
    print("B")
elif score >= 70:
    print("C")
```

```python
# MISTAKE 2: Wrong condition order
x = 15

# WRONG ORDER: less restrictive condition first
if x > 10:
    print("greater than 10")   # This always fires for x=15
elif x > 5:
    print("greater than 5")    # Never reached for x=15!

# For this example the order doesn't matter much, but consider:
x = 85
if x >= 60:              # More permissive first — WRONG ORDER!
    print("Passing")     # Always fires for x=85
elif x >= 80:
    print("Good")        # Never reached for x=85!
elif x >= 90:
    print("Excellent")
```
