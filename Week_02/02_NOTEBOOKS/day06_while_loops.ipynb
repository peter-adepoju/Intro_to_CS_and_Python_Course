## Section 2: The Three Parts — Initialize, Test, Update

```python
# Counting up by 2s
count = 0          # 1. INITIALIZE
while count < 20:  # 2. TEST
    print(count)
    count += 2     # 3. UPDATE
print("Done")
```

```python
# Counting DOWN
count = 10
while count > 0:
    print(count)
    count -= 1
print("Liftoff!")
```

## Section 3: Sentinel-Controlled Loops

Note: `input()` cells below will prompt you for text. Try entering a few
numbers, then enter -1 to stop.

```python
# Sentinel pattern: keep asking until the user enters -1
total = 0
num = int(input("Enter a number (or -1 to stop): "))   # priming read
while num != -1:
    total = total + num
    num = int(input("Enter a number (or -1 to stop): "))  # update read
print(f"Total: {total}")
```

## Section 4: Guided Practice — Building a Counting Loop From Scratch

Let's build a loop step by step. Goal: print all multiples of 7 between
7 and 70 (inclusive).

```python
# Step 1: initialize
n = 7

# Step 2 + 3: test and loop body with update
while n <= 70:
    print(n)
    n += 7

print("Done printing multiples of 7")
```

## Day 6 Wrap-Up

**Key things to remember from today:**

1. `while condition:` repeats as long as the condition stays True.
2. Every counting loop needs three parts: initialize, test, update.
3. Forgetting the update step causes an infinite loop.
4. The condition is checked BEFORE every pass — if False initially, the
   loop body never runs at all.
5. Sentinel-controlled loops need a priming read before the loop AND an
   update read inside the loop.
