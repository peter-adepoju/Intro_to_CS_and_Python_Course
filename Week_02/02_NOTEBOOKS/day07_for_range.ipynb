## Section 2: The Three Forms of `range()`

```python
# Form 1: range(stop) -- starts at 0
print("range(6):", list(range(6)))
```

```python
# Form 2: range(start, stop)
print("range(3, 9):", list(range(3, 9)))
```

```python
# Form 3: range(start, stop, step)
print("range(0, 20, 4):", list(range(0, 20, 4)))
print("range(20, 0, -4):", list(range(20, 0, -4)))
print("range(1, 10, 3):", list(range(1, 10, 3)))
```

> Note: `list(range(...))` is used above purely so you can SEE the
> values range() produces. Normally you'd just write `for i in range(...)`
> directly without wrapping it in `list()`.

## Section 3: The Accumulator Pattern

```python
# Sum accumulator -- starts at 0
total = 0
for i in range(1, 11):
    total = total + i
print("Sum 1-10:", total)
```

```python
# Product accumulator -- starts at 1!
factorial = 1
for i in range(1, 6):
    factorial = factorial * i
print("5! =", factorial)
```

```python
# String-building accumulator -- starts empty
result = ""
for i in range(5):
    result = result + str(i)
print(result)
```

```python
# Counting accumulator -- count multiples of 7 from 1 to 100
count = 0
for i in range(1, 101):
    if i % 7 == 0:
        count = count + 1
print("Multiples of 7 in 1-100:", count)
```

## Section 4: Iterating Over a String with `range(len(s))`

```python
s = "hello"
for i in range(len(s)):
    print(i, s[i])
```

```python
# Guided example: count vowels using range(len(...))
word = "programming"
vowel_count = 0
vowels = "aeiouAEIOU"

for i in range(len(word)):
    if word[i] in vowels:
        vowel_count = vowel_count + 1

print(f"'{word}' has {vowel_count} vowels")
```

## Day 7 Wrap-Up

**Key things to remember from today:**

1. `for variable in range(...):` repeats a known number of times.
2. `range(stop)`, `range(start, stop)`, `range(start, stop, step)` —
   `stop` is always EXCLUDED.
3. The accumulator pattern: initialize before the loop, update inside it.
   Sum starts at 0; product starts at 1.
4. `range(len(s))` gives you both the index AND, via `s[i]`, the character.
5. Use `for` when you know the number of repetitions; use `while` when
   you don't.
