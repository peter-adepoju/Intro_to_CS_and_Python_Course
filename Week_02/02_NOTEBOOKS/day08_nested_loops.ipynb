## Section 1: Direct String Iteration

```python
# The direct style -- no index needed
s = "hello"
for char in s:
    print(char)
```

```python
# Compare to the index-based style from yesterday
s = "hello"
for i in range(len(s)):
    print(s[i])

# Both print the same thing -- direct iteration is shorter when you
# don't need the index itself
```

```python
# When you DO need the index: comparing neighbors
s = "hello"
for i in range(len(s) - 1):   # note: -1, so we never go out of range
    if s[i] == s[i + 1]:
        print(f"Repeated letter at position {i}: {s[i]}")
```

## Section 3: Building Patterns

```python
# A simple grid
for row in range(4):
    line = ""
    for col in range(4):
        line = line + "* "
    print(line)
```

```python
# A triangle -- the inner range DEPENDS on the outer variable
for row in range(5):
    line = ""
    for col in range(row + 1):
        line = line + "*"
    print(line)
```

```python
# Multiplication table -- combines nested loops + accumulator + f-strings
for i in range(1, 6):
    row = ""
    for j in range(1, 6):
        product = i * j
        row = row + f"{product:>4}"
    print(row)
```

## Section 4: Nested Loops Over Two Strings

```python
# Finding matching characters between two strings
word1 = "abc"
word2 = "cde"

for char1 in word1:
    for char2 in word2:
        if char1 == char2:
            print(f"Match found: {char1}")
```

### ⚠️ Common Bug Demo: Reusing the Loop Variable

```python
# BUG: both loops use 'i' -- watch what goes wrong
print("Buggy version (shared variable name):")
for i in range(3):
    for i in range(2):       # this OVERWRITES the outer i!
        print("inner i:", i)
    print("after inner loop, i is:", i)   # never shows outer's intended value!

print()
print("Fixed version (distinct variable names):")
for i in range(3):
    for j in range(2):
        print("inner j:", j)
    print("outer i is still:", i)
```

## Day 8 Wrap-Up

**Key things to remember from today:**

1. `for char in s:` is the direct way to loop over string characters.
2. Use `range(len(s))` only when you actually need the index (e.g., to
   compare to a neighbor).
3. In nested loops, the INNER loop completes entirely for every single
   pass of the OUTER loop.
4. Total iterations = outer_count × inner_count.
5. Always use DIFFERENT variable names for nested loops (i/j, row/col).
6. The inner loop's range() can depend on the outer loop's current value.
