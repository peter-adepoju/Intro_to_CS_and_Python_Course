## Section 1: Guess-and-Check Square Root

```python
x = int(input("Enter a non-negative integer: "))
guess = 0
while guess ** 2 < x:
    guess += 1

if guess ** 2 == x:
    print(f"The square root of {x} is {guess}")
else:
    print(f"{x} is not a perfect square")
```

```python
# Handling negative numbers gracefully
x = int(input("Enter an integer: "))
neg_flag = False
if x < 0:
    neg_flag = True

guess = 0
while guess ** 2 < x:
    guess += 1

if guess ** 2 == x:
    print(f"The square root of {x} is {guess}")
else:
    print(f"{x} is not a perfect square")
    if neg_flag:
        print(f"(Note: {x} is negative, so it has no real square root)")
```

## Section 2: Cube Roots (Including Negatives)

```python
cube = int(input("Enter an integer: "))
found = False
for guess in range(abs(cube) + 1):
    if guess ** 3 == abs(cube):
        found = True
        if cube < 0:
            guess = -guess
        print(f"Cube root of {cube} is {guess}")
        break
if not found:
    print(f"{cube} is not a perfect cube")
```

## Section 3: Exhaustive Enumeration — The Ticket Problem

```python
# Alyssa, Ben, and Cindy together sold 10 tickets.
# Ben sold 2 fewer than Alyssa. Cindy sold twice as many as Alyssa.

for alyssa in range(11):
    for ben in range(11):
        for cindy in range(11):
            sells_total = (alyssa + ben + cindy == 10)
            two_less = (ben == alyssa - 2)
            twice = (cindy == 2 * alyssa)
            if sells_total and two_less and twice:
                print(f"Alyssa sold {alyssa} tickets")
                print(f"Ben sold {ben} tickets")
                print(f"Cindy sold {cindy} tickets")
```

```python
# The SAME problem, solved with a single, smarter loop
# (no need to search for ben and cindy -- compute them directly!)

for alyssa in range(11):
    ben = alyssa - 2
    cindy = alyssa * 2
    if alyssa + ben + cindy == 10:
        print(f"Alyssa sold {alyssa} tickets")
        print(f"Ben sold {ben} tickets")
        print(f"Cindy sold {cindy} tickets")

# Notice: same answer, vastly fewer iterations (11 vs 1331)
```

## Section 4: Floating-Point Accumulation Error

```python
# 0.1 added ten times does NOT give exactly 1.0
x = 0
for i in range(10):
    x += 0.1
print(x)
print(x == 1.0)
```

```python
# But 0.125 (which IS exactly representable in binary) works perfectly
x = 0
for i in range(10):
    x += 0.125
print(x)
print(x == 1.25)
```

```python
# The fix: "close enough" comparison instead of exact equality
x = 0
for i in range(10):
    x += 0.1
print(abs(x - 1.0) < 0.0001)   # True -- this is the correct way to compare floats
```

## Section 5: Decimal to Binary Conversion

```python
num = int(input("Enter a positive integer: "))
result = ''
if num == 0:
    result = '0'
while num > 0:
    result = str(num % 2) + result
    num = num // 2
print(f"Binary representation: {result}")
```

## Day 9 Wrap-Up

**Key things to remember from today:**

1. Guess-and-check loops systematically test increasing values until the
   answer is found or surpassed.
2. Always consider edge cases: zero, negative numbers, very large inputs.
3. Exhaustive enumeration with nested loops always works but can be slow —
   look for ways to compute unknowns directly instead of searching for them.
4. Floating-point arithmetic accumulates small errors — never use `==` to
   compare computed floats; use `abs(a - b) < tolerance` instead.
