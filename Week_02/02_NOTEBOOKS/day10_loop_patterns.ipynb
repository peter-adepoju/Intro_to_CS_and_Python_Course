## Section 1: `break`

```python
for i in range(10):
    if i == 5:
        break
    print(i)
print("Loop exited")
```

```python
# break only exits the INNERMOST loop
for i in range(3):
    for j in range(3):
        if j == 1:
            break
        print(i, j)
# Outer loop is unaffected -- it still runs all 3 times
```

## Section 2: `continue`

```python
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
# Only odd numbers print -- continue skips just that one pass
```

```python
# break vs continue side by side
print("break:")
for i in range(6):
    if i == 3:
        break
    print(i)

print("continue:")
for i in range(6):
    if i == 3:
        continue
    print(i)
```

## Section 3: Boolean Flags — Search and Report

## Section 4: The Pattern Catalog — Six Patterns in Action

```python
# Pattern 1: Counting
count = 0
for i in range(1, 51):
    if i % 6 == 0:
        count += 1
print("Counting pattern:", count)
```

```python
# Pattern 2: Accumulating (sum)
total = 0
for i in range(1, 51):
    if i % 6 == 0:
        total += i
print("Accumulating pattern:", total)
```

```python
# Pattern 3: Building a string
s = "hello world"
result = ""
for char in s:
    if char != " ":
        result += char
print("Building pattern:", result)
```

```python
# Pattern 4: Search-and-report
s = "the quick brown fox"
found_flag = False
for char in s:
    if char == 'z':
        found_flag = True
        break
print("Search-and-report pattern: contains z?", found_flag)
```

```python
# Pattern 5: Finding an extreme value
numbers = [12, 45, 3, 67, 21, 9]
largest = numbers[0]    # IMPORTANT: start with the first actual element!
for num in numbers:
    if num > largest:
        largest = num
print("Extreme-value pattern (max):", largest)
```

```python
# Pattern 6: ALL / ANY
numbers = [4, 8, 12, 16, 20]

all_even = True
for num in numbers:
    if num % 2 != 0:
        all_even = False
        break
print("ALL pattern (all even?):", all_even)

any_negative = False
for num in numbers:
    if num < 0:
        any_negative = True
        break
print("ANY pattern (any negative?):", any_negative)
```

## Section 5: Worked Example — Combining Three Patterns

```python
# First repeated character: combines accumulator + search-and-report
s = "discover"
seen = ""
repeated_char = None
found_flag = False

for char in s:
    if char in seen:
        repeated_char = char
        found_flag = True
        break
    seen += char

if found_flag:
    print(f"First repeated character: {repeated_char}")
else:
    print("No repeated characters found.")

# Try changing s to "hello" and re-running -- the second 'l' should trigger it
```
