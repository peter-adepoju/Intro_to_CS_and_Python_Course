## Section 1: The Problem Functions Solve

```python
# WITHOUT a function -- repeated logic
n1 = 14
if n1 % 2 == 0:
    print(f"{n1} is even")

n2 = 27
if n2 % 2 == 0:
    print(f"{n2} is even")

n3 = 8
if n3 % 2 == 0:
    print(f"{n3} is even")
```

```python
# WITH a function -- logic written once
def is_even(n):
    return n % 2 == 0

print(is_even(14))
print(is_even(27))
print(is_even(8))
```

## Section 2: Defining vs. Calling

```python
# Defining a function does NOT run it
def greet():
    print("Hello! Welcome to the course.")

# Nothing printed yet -- we only DEFINED greet, we haven't CALLED it
```

```python
# NOW we call it -- this actually runs the body
greet()
greet()
greet()
```

### ⚠️ Common Mistake: Forgetting Parentheses

```python
greet     # This does NOT call the function -- just refers to it
          # Notice: nothing prints, and Jupyter shows a function object,
          # not the result of running it
```

## Section 3: Functions With Parameters

```python
def square(x):
    return x * x

print(square(5))
print(square(-3))
print(square(0))
```

```python
# Multiple parameters -- matched by POSITION
def add(a, b):
    return a + b

print(add(3, 4))
print(add(10, -2))
```

```python
def describe_rectangle(width, height):
    area = width * height
    perimeter = 2 * (width + height)
    print(f"Width: {width}, Height: {height}")
    print(f"Area: {area}, Perimeter: {perimeter}")

describe_rectangle(5, 3)
```

## Section 4: Define-Before-Call Order Matters

```python
# This will raise a NameError -- the function isn't defined yet
try:
    not_yet_defined()
except NameError as e:
    print(f"NameError: {e}")

def not_yet_defined():
    print("Now I exist!")

# NOW it works, since the def has run
not_yet_defined()
```

## Day 11 Wrap-Up

**Key things to remember from today:**

1. Functions exist to avoid repeated code and organize logic.
2. `def name(params):` DEFINES a function — it does not run the body.
3. `name(args)` CALLS the function — this actually runs it.
4. A function must be defined before any line that calls it executes.
5. Parameters are matched to arguments by position, by default.
