## Section 1: Positional Argument Matching

```python
def describe_pet(name, animal_type, age):
    print(f"{name} is a {age}-year-old {animal_type}")

describe_pet("Rex", "dog", 3)
```

```python
# Swap the order -- still 'works', but the MEANING is wrong!
describe_pet("dog", "Rex", 3)
# No error -- but this is clearly not what we intended
```

## Section 2: Default Parameter Values

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Alice")                   # uses the default
greet("Bob", "Hi")                 # overrides the default
greet("Carol", greeting="Hey")     # overrides, using keyword form
```

```python
def make_separator(length=40, character="-"):
    return character * length

print(make_separator())
print(make_separator(10))
print(make_separator(10, "="))
```

### ⚠️ Default Parameters Must Come AFTER Non-Default Ones

```python
# This is a SyntaxError -- uncomment to see it for yourself:
# def f(a=10, b):
#     return a + b

# Correct order:
def f(b, a=10):
    return a + b

print(f(5))        # uses default a=10: 15
print(f(5, 20))     # overrides a: 25
```

## Section 3: Keyword Arguments

```python
def describe_pet(name, animal_type, age):
    print(f"{name} is a {age}-year-old {animal_type}")

# All three of these produce the SAME result:
describe_pet("Rex", "dog", 3)
describe_pet(name="Rex", animal_type="dog", age=3)
describe_pet(age=3, name="Rex", animal_type="dog")   # order doesn't matter!
```

```python
# Mixing positional and keyword -- positional MUST come first
describe_pet("Rex", age=3, animal_type="dog")   # OK
```

## Section 4: A Complete Example

```python
def simple_interest(principal, rate, years=1):
    """Returns the simple interest earned.
    principal: the starting amount (a number)
    rate: the annual interest rate as a decimal (e.g., 0.05 for 5%)
    years: the number of years (defaults to 1)
    """
    return principal * rate * years

print(simple_interest(1000, 0.05))          # uses default years=1
print(simple_interest(1000, 0.05, 10))      # 10 years
print(simple_interest(1000, 0.05, years=3)) # keyword form
```

## Section 5: Argument Count Mismatches

```python
def add(a, b):
    return a + b

try:
    add(5)
except TypeError as e:
    print(f"TypeError: {e}")

try:
    add(5, 3, 1)
except TypeError as e:
    print(f"TypeError: {e}")
```

## Day 12 Wrap-Up

**Key things to remember from today:**

1. Parameters (definition) vs. arguments (the actual call values).
2. Positional matching is the default — order matters, and mistakes are
   silent (no error), so be careful.
3. Default parameter values must come after non-default ones.
4. Keyword arguments are order-independent and self-documenting.
5. Positional arguments must come before keyword arguments in a call.
