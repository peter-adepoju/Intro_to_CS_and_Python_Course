## Section 1: `print` vs. `return`

```python
def add_v1(a, b):
    print(a + b)     # only DISPLAYS

def add_v2(a, b):
    return a + b      # sends the result BACK

result1 = add_v1(3, 4)   # prints 7 to the screen
result2 = add_v2(3, 4)   # does NOT print anything

print("result1:", result1)   # None!
print("result2:", result2)   # 7
```

```python
def square_print(x):
    print(x * x)

def square_return(x):
    return x * x

# With return version: this works perfectly
total = square_return(3) + square_return(4)
print("total:", total)

# With print-only version: this would ERROR (uncomment to see)
# total2 = square_print(3) + square_print(4)
```

## Section 3: `return` Exits Immediately

```python
def classify_number(n):
    if n < 0:
        return "negative"
    if n == 0:
        return "zero"
    return "positive"

print(classify_number(-5))
print(classify_number(0))
print(classify_number(7))
```

```python
def grade_letter(score):
    if score >= 90:
        return "A"
    if score >= 80:
        return "B"
    if score >= 70:
        return "C"
    if score >= 60:
        return "D"
    return "F"

for test_score in [95, 82, 71, 60, 45]:
    print(test_score, "->", grade_letter(test_score))
```

## Section 4: Returning Multiple Values

```python
def min_and_max(a, b, c):
    smallest = a
    if b < smallest:
        smallest = b
    if c < smallest:
        smallest = c

    largest = a
    if b > largest:
        largest = b
    if c > largest:
        largest = c

    return smallest, largest

low, high = min_and_max(7, 2, 9)
print(low, high)
```

## Section 5: Local Scope

```python
def compute_area(width, height):
    area = width * height   # LOCAL to this function
    return area

result = compute_area(5, 3)
print(result)

try:
    print(area)   # NameError! 'area' only existed INSIDE compute_area
except NameError as e:
    print(f"NameError: {e}")
```

```python
# Each call gets a FRESH, independent local scope
def add_one(x):
    x = x + 1
    return x

a = add_one(5)
b = add_one(100)
print(a, b)   # 6 101 -- completely independent calls
```

## Section 6: Global Scope and Its Pitfalls

```python
# Functions CAN read global variables
discount_rate = 0.1

def apply_discount(price):
    return price * (1 - discount_rate)

print(apply_discount(100))
```

```python
# But assigning to a name inside a function makes it LOCAL for the
# whole function body -- this raises UnboundLocalError:
counter = 0

def broken():
    print(counter)     # tries to read 'counter' -- but Python has
                         # already decided counter is LOCAL because of
                         # the next line!
    counter = counter + 1

try:
    broken()
except UnboundLocalError as e:
    print(f"UnboundLocalError: {e}")
```

```python
# The 'global' keyword (works, but NOT the preferred style in this course)
counter = 0

def increment():
    global counter
    counter = counter + 1

increment()
increment()
print(counter)
```

```python
# PREFERRED PATTERN: pass in current state, return new state
counter = 0

def increment(current_value):
    return current_value + 1

counter = increment(counter)
counter = increment(counter)
print(counter)   # same result, no 'global' needed
```

## Day 13 Wrap-Up

**Key things to remember from today:**

1. `print` displays for humans; `return` sends a usable value to the caller.
2. A function with no `return` gives back `None`.
3. `return` exits the function immediately — code after it never runs.
4. You can return multiple values: `return a, b` then `x, y = f(...)`.
5. Local scope: variables inside a function are invisible outside it.
6. Avoid `global` — prefer passing in state and returning new state.
