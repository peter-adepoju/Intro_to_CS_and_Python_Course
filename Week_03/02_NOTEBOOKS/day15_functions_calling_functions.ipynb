```python
def square(x):
    return x * x

def sum_of_squares(a, b):
    return square(a) + square(b)

print(sum_of_squares(3, 4))   # 9 + 16 = 25
```

## Section 2: Tracing a Multi-Function Program

```python
def add_tax(price, rate):
    return price * (1 + rate)

def apply_discount(price, discount):
    return price * (1 - discount)

def final_price(price, discount, tax_rate):
    discounted = apply_discount(price, discount)
    final = add_tax(discounted, tax_rate)
    return final

result = final_price(100, 0.1, 0.08)
print(result)

# Trace it yourself before running:
# 1. apply_discount(100, 0.1) = ?
# 2. add_tax(that result, 0.08) = ?
```

## Section 3: Bisection Search — A Faster Approximation

```python
def bisection_sqrt(x, epsilon=0.01):
    """
    Assumes: x is a non-negative number, epsilon is a small positive number
    Returns: an approximation of the square root of x, accurate to
             within epsilon
    """
    low = 0.0
    high = max(x, 1.0)
    guess = (low + high) / 2

    while abs(guess ** 2 - x) >= epsilon:
        if guess ** 2 < x:
            low = guess
        else:
            high = guess
        guess = (low + high) / 2

    return guess

print(bisection_sqrt(100000))   # close to 316.23
print(bisection_sqrt(0.25))      # close to 0.5
print(bisection_sqrt(2))          # close to 1.414...
```

### ⏱️ Speed Comparison: Bisection vs. Guess-and-Check

⚠️ **Important:** only run this comparison with `x = 1000` or smaller —
larger values can make the guess-and-check version run for a very long
time, or even fail to terminate at all (see the textbook's warning box
in section 15.4 about why).

```python
def guess_and_check_sqrt(x, epsilon=0.01, increment=0.0001):
    """Slow approach from Week 2, for comparison."""
    guess = 0.0
    num_guesses = 0
    while abs(guess ** 2 - x) >= epsilon:
        guess += increment
        num_guesses += 1
    return guess, num_guesses

def bisection_sqrt_counted(x, epsilon=0.01):
    """Same as bisection_sqrt, but also counts steps for comparison."""
    low = 0.0
    high = max(x, 1.0)
    guess = (low + high) / 2
    num_guesses = 0
    while abs(guess ** 2 - x) >= epsilon:
        if guess ** 2 < x:
            low = guess
        else:
            high = guess
        guess = (low + high) / 2
        num_guesses += 1
    return guess, num_guesses

x = 1000   # SAFE value -- do not increase this significantly, see warning above
_, slow_count = guess_and_check_sqrt(x)
_, fast_count = bisection_sqrt_counted(x)
print(f"Guess-and-check: {slow_count} guesses")
print(f"Bisection search: {fast_count} guesses")
```

## Section 4: A Gentle First Look at Recursion

```python
def countdown(n):
    """
    Assumes: n is a non-negative integer
    Prints the numbers from n down to 0, each on its own line.
    """
    if n == 0:
        print("Liftoff!")
    else:
        print(n)
        countdown(n - 1)   # the function calls ITSELF

countdown(3)
```
