## Section 1: The Core Idea

````python
# Factorial: n! = n * (n-1) * (n-2) * ... * 1, and 0! = 1
def factorial(n):
    """
    Assumes: n is a non-negative integer
    Returns: n! (n factorial)
    """
    if n == 0:               # BASE CASE: directly answerable
        return 1
    else:                    # RECURSIVE CASE: smaller problem
        return n * factorial(n - 1)

print(factorial(0))
print(factorial(1))
print(factorial(5))
````

## Section 2: Recursive vs. Iterative — Same Problem, Two Tools

````python
# Recursive sum_to_n
def sum_to_n(n):
    """
    Assumes: n is a non-negative integer
    Returns: sum of all integers from 1 to n
    """
    if n == 0:
        return 0
    else:
        return n + sum_to_n(n - 1)

# Iterative sum_to_n (from Week 2)
def sum_to_n_iterative(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

print(sum_to_n(5))           # 15
print(sum_to_n_iterative(5)) # 15 -- both give the same answer

# Verify they always agree
for n in range(10):
    assert sum_to_n(n) == sum_to_n_iterative(n)
print("All match for n=0..9")
````

## Section 3: Designing Recursion Step by Step

````python
# count_digits: how many digits are in a non-negative integer?
#
# Step 1 -- base case: single digit number (n < 10) -> answer is 1
# Step 2 -- recursive case: remove last digit with n // 10, add 1
# Step 3 -- does it shrink? Yes -- n // 10 < n for n >= 10

def count_digits(n):
    """
    Assumes: n is a non-negative integer
    Returns: the number of digits in n
    """
    if n < 10:
        return 1
    else:
        return 1 + count_digits(n // 10)

print(count_digits(7))       # 1
print(count_digits(42))      # 2
print(count_digits(123456))  # 6
````

## Section 4: "Trusting the Recursion"

````python
# Instead of tracing every level, ask:
# - Is the base case correct?
# - IF the smaller recursive call were correct, is MY computation right?

def power(base, exponent):
    """
    Assumes: exponent is a non-negative integer
    Returns: base ** exponent (using only recursion and multiplication)
    """
    if exponent == 0:            # Base case: anything^0 = 1
        return 1
    return base * power(base, exponent - 1)   # Trust that power(b, e-1)
                                               # correctly returns b^(e-1)
print(power(2, 10))  # 1024
print(power(3, 4))   # 81
print(power(5, 0))   # 1
````
