```python
# Recap: correct factorial from Day 16
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

## Section 1: What Happens Without a Base Case

```python
import sys
print("Python recursion limit:", sys.getrecursionlimit())

def broken_countdown(n):
    print(n)
    broken_countdown(n - 1)   # no base case!

try:
    broken_countdown(5)
except RecursionError:
    print("RecursionError -- no base case, recursed until Python stopped it")
```

## Section 2: Base Case Present But Never Reached

```python
def broken_same_n(n):
    if n == 0:
        return 1
    return n * broken_same_n(n)   # BUG: passes n, not n-1!

try:
    broken_same_n(5)
except RecursionError:
    print("RecursionError -- base case exists but is never reached")
```

## Section 3: Skipping Over the Base Case

```python
# Steps of 2 can skip over n == 0
def broken_skip(n):
    if n == 0:
        print("Done!")
    else:
        print(n)
        broken_skip(n - 2)   # BUG: odd n never hits exactly 0

print("Even start (works):")
broken_skip(4)

print("\nOdd start (RecursionError):")
try:
    broken_skip(5)
except RecursionError:
    print("RecursionError -- 5, 3, 1, -1, -3, ... never reaches 0")

# Fixed version
def fixed_countdown(n):
    if n <= 0:         # catches ANY value at or below 0
        print("Done!")
    else:
        print(n)
        fixed_countdown(n - 2)

print("\nFixed odd start:")
fixed_countdown(5)
```

## Section 4: Multiple Base Cases — Fibonacci Preview

```python
def fibonacci(n):
    """
    Assumes: n is a non-negative integer
    Returns: the nth Fibonacci number
    """
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(10):
    print(f"fib({i}) = {fibonacci(i)}")
```

## Day 17 Wrap-Up

**Key things to remember:**
1. Every recursive function needs a base case that is both present AND reachable.
2. RecursionError usually means a missing or unreachable base case — not a need for more recursion.
3. The recursive argument must shrink on EVERY call without exception.
4. Steps bigger than 1 can skip over `== 0`; use `<= 0` when appropriate.
