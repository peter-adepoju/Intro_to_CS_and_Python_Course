## Section 1: Winding and Unwinding Visualized

```python
# This version prints WHEN each multiplication happens,
# making winding/unwinding visible
def factorial_traced(n):
    if n == 0:
        print(f"  Base case: factorial(0) = 1")
        return 1
    print(f"  Winding:   factorial({n}) waiting for factorial({n-1})")
    result = n * factorial_traced(n - 1)
    print(f"  Unwinding: factorial({n}) = {n} * result = {result}")
    return result

print("factorial_traced(4):")
ans = factorial_traced(4)
print("Final answer:", ans)
```

## Section 2: Tracing Winding vs Unwinding for Print-Only Recursion

```python
# Notice that print BEFORE the recursive call happens during winding,
# and print AFTER happens during unwinding
def trace_demo(n):
    if n == 0:
        print("base case reached")
        return
    print(f"winding: about to recurse with n={n}")
    trace_demo(n - 1)
    print(f"unwinding: back at n={n}")

# PREDICT the complete output BEFORE running:
trace_demo(3)
```

## Section 3: State Table Tracing — power(2, 4)

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    return base * power(base, exponent - 1)

# Trace power(2, 4) yourself with a state table:
# | Call           | base | exponent | waiting on      | returns |
# |----------------|------|----------|-----------------|---------|
# | power(2, 4)    |  2   |    4     | power(2, 3)     |  2*?    |
# | power(2, 3)    |  2   |    3     | power(2, 2)     |  2*?    |
# | ...            |      |          |                 |         |

print(power(2, 4))   # 16
print(power(3, 4))   # 81
```

## Section 4: String Recursion and Call Tree Preview

```python
def reverse_string(s):
    """
    Assumes: s is a string
    Returns: s reversed
    """
    if len(s) <= 1:
        return s
    return reverse_string(s[1:]) + s[0]

# PREDICT each result before running:
print(reverse_string("abc"))    # Prediction: ____
print(reverse_string("hi"))     # Prediction: ____
print(reverse_string("a"))      # Prediction: ____
print(reverse_string(""))        # Prediction: ____
```

```python
# Recursion depth: Python's limit is 1000
import sys
print("Limit:", sys.getrecursionlimit())

def count_up(n, limit):
    if n > limit:
        return
    count_up(n + 1, limit)

try:
    count_up(1, 2000)
except RecursionError:
    print("RecursionError at depth > 1000")
```

## Day 18 Wrap-Up

**Key things to remember:**
1. Winding: calls go deeper; no computation yet (for value-returning functions).
2. Unwinding: base case returns, results travel back up.
3. Each call has its own independent local scope -- even calls to the same function.
4. Code AFTER a recursive call still runs during unwinding.
