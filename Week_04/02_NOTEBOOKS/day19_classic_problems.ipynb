## Section 1: Fibonacci — Naive Version and Its Cost

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

for i in range(11):
    print(f"fib({i}) = {fibonacci(i)}")
```

```python
# Count how many calls naive fibonacci makes
call_count = 0

def fibonacci_counted(n):
    global call_count
    call_count += 1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci_counted(n - 1) + fibonacci_counted(n - 2)

for n in [10, 20, 30]:
    call_count = 0
    result = fibonacci_counted(n)
    print(f"fib({n}) = {result}, total calls = {call_count}")
```

## Section 2: Efficient Fibonacci

```python
import time

def fib_efficient(n, a=0, b=1):
    """
    Assumes: n is a non-negative integer
    Returns: the nth Fibonacci number in O(n) recursive calls
    """
    if n == 0:
        return a
    return fib_efficient(n - 1, b, a + b)

# Compare timing
start = time.time()
r1 = fibonacci(32)
slow_time = time.time() - start

start = time.time()
r2 = fib_efficient(32)
fast_time = time.time() - start

print(f"Naive fib(32)    = {r1}, time = {slow_time:.3f}s")
print(f"Efficient fib(32)= {r2}, time = {fast_time:.6f}s")
print()
print("fib_efficient(100):", fib_efficient(100))  # instant!
```

## Section 3: Towers of Hanoi

```python
def hanoi(num_disks, source, target, auxiliary):
    """
    Assumes: num_disks is a non-negative integer
    Prints each move and returns the total number of moves.
    """
    if num_disks == 0:
        return 0
    moves_before = hanoi(num_disks - 1, source, auxiliary, target)
    print(f"Move disk from {source} to {target}")
    moves_after = hanoi(num_disks - 1, auxiliary, target, source)
    return moves_before + 1 + moves_after

print("hanoi(3, A, C, B):")
total = hanoi(3, 'A', 'C', 'B')
print(f"Total moves: {total}  (expected: {2**3 - 1})")
```

```python
# Verify the 2^n - 1 formula
import io, contextlib

for n in range(1, 7):
    buf = io.StringIO()
    with contextlib.redirect_stdout(buf):
        total = hanoi(n, 'A', 'C', 'B')
    expected = 2**n - 1
    print(f"n={n}: {total} moves (2^{n}-1={expected}) -- {'✓' if total == expected else '✗'}")
```

## Section 4: Recursive String Processing

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])

def contains_char(s, target, index=0):
    if index >= len(s):
        return False
    if s[index] == target:
        return True
    return contains_char(s, target, index + 1)

# PREDICT each result before running:
print(is_palindrome("racecar"))   # Prediction: ____
print(is_palindrome("hello"))     # Prediction: ____
print(is_palindrome("level"))     # Prediction: ____

print(contains_char("hello", "l"))  # Prediction: ____
print(contains_char("hello", "z"))  # Prediction: ____
```
