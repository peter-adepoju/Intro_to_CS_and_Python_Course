## Section 2: Recursion vs. Iteration Side by Side

````python
# Factorial: both forms side by side
def factorial_recursive(n):
    if n == 0:
        return 1
    return n * factorial_recursive(n - 1)

def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

for n in range(8):
    r = factorial_recursive(n)
    i = factorial_iterative(n)
    match = "✓" if r == i else "✗"
    print(f"{n}!: recursive={r}, iterative={i} {match}")
````

````python
# sum_to_n: both forms
def sum_recursive(n):
    if n == 0:
        return 0
    return n + sum_recursive(n - 1)

def sum_iterative(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

for n in [0, 5, 10, 100]:
    assert sum_recursive(n) == sum_iterative(n)
print("All match")
````

## Section 3: When Recursion Wins

````python
# Towers of Hanoi -- 5 clean lines vs. a much messier iterative version
import io, contextlib

def hanoi(num_disks, source, target, auxiliary):
    if num_disks == 0:
        return 0
    before = hanoi(num_disks - 1, source, auxiliary, target)
    print(f"Move disk from {source} to {target}")
    after = hanoi(num_disks - 1, auxiliary, target, source)
    return before + 1 + after

buf = io.StringIO()
with contextlib.redirect_stdout(buf):
    total = hanoi(4, 'A', 'C', 'B')
print(f"hanoi(4): {total} moves (expected {2**4-1})")
````

## Section 4: Converting Between Forms

````python
# Recursive form (to convert to iterative)
def countdown_recursive(n):
    if n < 0:
        return
    print(n)
    countdown_recursive(n - 1)

# Iterative equivalent
def countdown_iterative(n):
    while n >= 0:
        print(n)
        n -= 1

print("Recursive:")
countdown_recursive(5)
print("Iterative:")
countdown_iterative(5)
````
